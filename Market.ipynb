```python
# Analyzing and drawing the Graph of 'Open,Adj Close' at a Market dataset
# Detecting Noises & Missing values and Cleaning them
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv ('E:\Momeni-Machne Learning class momeni 21 01 1402\data science\datasets/Market.csv')
df=pd.DataFrame(data)
df
```

```python
# We need Just These columns 'NYA', so delete other columns
df1 =  df[df['Index'] == 'NYA'] 
df1
```

```python
# Removing the last column because it is all zero
df2 = df1.drop(columns = 'Volume')
df2
```

```python
# Converting all floats to integers 
df3 = round(df2)
df3
```

```python
# Sorting based on Date
df3.sort_values(by = 'Date')
```

```python
# Noise and error detection
df3.isnull().sum()
```

```python
# A Toral Cleaning of Mising Value 
df4 = df3.dropna()
df4
```

#  Identification and cleaning of noise of 'Open_Adj Close'

```python
plt.figure(figsize =(10,8))
plt.scatter (df4['Open'], df4['Adj Close'])
plt.title('Open_Adj Close', fontsize= 30)
plt.xlabel('Open', fontsize=30)
plt.xticks(range(0,17000,1000))
plt.xticks(rotation=90)
plt.ylabel('Adj Close', fontsize = 30)
plt.yticks(range(0,17000,1000))
plt.legend(['show Open'], loc = 'best')
plt.grid()
plt.show()
```

```python
# Below noise
Noise = df4[(df4['Open']>5000) & (df4['Adj Close']<1000)]
Noise
```

```python
# second way 
Noise = df4[df4['Open']>5000]
Noise
```

```python
# upper noise
Noise2 = df4[(df4['Open']<1000) & (df4['Adj Close']>5000)]
Noise2
```

```python
#  second way
Noise2 = df4[df4['Adj Close']>5000] 
Noise2
```

```python
# Deleting the detected noise
df4.drop(index = [831,852], inplace = True)
df4
```

#   Redraw the graph and check the removed noise

```python
plt.figure(figsize =(10,8))
plt.scatter (df4['Open'], df4['Adj Close'])
plt.title('Open_Adj Close', fontsize= 30)
plt.xlabel('Open', fontsize=30)
plt.xticks(range(0,17000,1000))
plt.xticks(rotation=90)
plt.ylabel('Adj Close', fontsize = 30)
plt.yticks(range(0,17000,1000))
plt.legend(['show Open'], loc = 'best')
plt.grid()
plt.show()
```
